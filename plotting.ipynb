# ZTF-TESS Plotting

```python
import pandas as pd

from preprocess import preprocess_ztf_tess, display_all_passbands, format_title, find_max_light, preprocess_tess
import os
import warnings
import matplotlib.pyplot as plt

params = {
        "norm": True,
        "to_bin": True,
        "bin_interval": "0.5D",
        "time_scale": "trigger",
        "convert_to_mag": True,
        "sub_bg_model": False,
        "remove_extinction": True
    }
ztf_data = os.listdir("TESS_data/ztf_data/")

def plot_curve(filename, save=False, save_dir="./TESS_data/ztf_tess_plots/all/", save_name="test"):
    xlim = [-30,30]
    fig, axs = plt.subplots(1, 2, figsize=(12, 3))

    # plot normalized
    params['norm'] = True
    df, meta = preprocess_ztf_tess(filename, params)
    if df is not None:
        display_all_passbands(df, meta, flip=True, xlim=xlim, ax=axs[0], title_info="Normalized")

        # plot unnormalized
        params['norm'] = False
        df_nonorm, meta = preprocess_ztf_tess(filename, params)
        display_all_passbands(df_nonorm, meta, flip=True, xlim=xlim, ax=axs[1], title_info="Unnormalized")

        fig.suptitle(format_title(meta), y=1.35)
        if save:
            fig.savefig(save_dir + save_name, bbox_inches="tight")
    if save:
        plt.close(fig)
    return df_nonorm
```

# TESS-Plotting

```python
import os
import matplotlib.pyplot as plt
from preprocess import display_curve, format_title, preprocess_tess
import warnings
import json
import pandas as pd

light_curves = os.listdir("./TESS_data/ztf_data")
save_dir="./TESS_data/tess_unprocessed_plots/all/"
not_found = []
saved = []
xlim = [-30,30]
params = {
        "norm": False,
        "to_bin": False,
        "bin_interval": "0.5D",
        "time_scale": "trigger",
        "sub_bg_model": False,
        "convert_to_mag": False,
        "remove_extinction": True
    }

def save_tess_plot(tess_curve_name):
    tess_file_name = f'lc_{tess_curve_name}_cleaned'
    params['to_bin'] = False
    df, meta = preprocess_tess(tess_file_name, params)
    if df is not None:
        fig, ax = plt.subplots()
        display_curve(df, meta, "cts", "e_cts", ax=ax, is_mag=False, alpha=0.2)
        params['to_bin'] = True

        df1, meta = preprocess_tess(tess_file_name, params)
        display_curve(df1, meta, "cts", "e_cts", ax=ax, is_mag=False, alpha=1.0, color="blue", max_color="blue", label="binned")

        ax.set_title(format_title(meta))
        ax.set_xlim(xlim)
        fig.savefig(save_dir + tess_curve_name, bbox_inches="tight")
        saved.append(tess_curve_name)
        # plt.close(fig)
        return df, df1, meta
    else:
        not_found.append(tess_curve_name)

def save_all(curves):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        for curve_name in curves:
            tess_name = curve_name.split("_")[0]
            save_tess_plot(tess_name)
```

# Test median window filtering

```python
import scipy.signal as ss

window_size = 3

_, df, meta = save_tess_plot("2020fcw")
df_rolling_center = df.copy()
df_rolling_center["cts"] = df_rolling_center["cts"].rolling(window_size, center=True).median()
ax = display_curve(df_rolling_center, meta, "cts", "e_cts", is_mag=False, alpha=1.0, color="blue", max_color="blue")
ax.set_xlim([-30,30])

df_scipy = df.copy()
df_scipy["cts"] = pd.Series(ss.medfilt(df_scipy["cts"].to_numpy(), 3), index=df.index)
ax = display_curve(df_scipy, meta, "cts", "e_cts", is_mag=False, alpha=1.0, color="blue", max_color="blue")
ax.set_xlim([-30,30])
```

```python
save_all(light_curves)
```

## save all curves with desired labels

```python
import shutil
import pandas as pd

def remove_dir_files(dir):
    for f in os.listdir(dir):
        os.remove(os.path.join(dir, f))

labels = pd.read_csv("./TESS_data/curve_labels.csv")
tess_dir = f"./TESS_data/tess_unprocessed_plots/"
good_dir = tess_dir + "good"
maybe_dir = tess_dir + "maybe"
great_dir = tess_dir + "great"

remove_dir_files(good_dir)
remove_dir_files(maybe_dir)
remove_dir_files(great_dir)


good_labels = labels[(labels["tess_good"] == True)]["curve_name"]
maybe_labels = labels[(labels["tess_maybe"] == True)]["curve_name"]
great_labels = labels[(labels["tess_great"] == True)]["curve_name"]

for file in maybe_labels:
    shutil.copy(f"{tess_dir}/all/{file}.png", maybe_dir)

for file in good_labels:
    shutil.copy(f"{tess_dir}all/{file}.png", good_dir)

for file in great_labels:
    shutil.copy(f"{tess_dir}all/{file}.png", great_dir)
```
